## O Danych

Zbiór danych zawiera informacje o trzech gatunkach irysów: Iris setosa, Iris versicolor, i Iris virginica.

Dane obejmują pomiary czterech cech: długość i szerokość działki kielicha oraz długość i szerokość płatka.

Każdy wiersz w zbiorze danych reprezentuje pojedynczy kwiat, a wartości pomiarów są podane w centymetrach.

Zbiór składa się z 150 próbek, po 50 dla każdego gatunku, i jest szeroko stosowany jako podstawowy zbiór danych do

* **długość kielicha (sepal length)** - Długość kielicha w cm

##### *tu pewnie najważniejsza jest klasa irysa - sprawdzę, czy wartości w poszczególnych kolumnach zależą od klasy i dlaczego jest pewna prawidłowość w wielkości płatków, że najwięcej jest najmniejszych*

##### *mogłem to robić po kolei, ale posatwnowiłem poprosić o pomoc `ChatGPT`*

```python
#mogłem to robić po kolei, ale posatwnowiłem poprosić o pomoc ChatGPT
for col in ['długość kielicha (sepal length)', 
            'szerokość kielicha (sepal width)', 
            'długość płatka (petal length)', 
            'szerokość płatka (petal width)']:
    df.plot(kind='scatter', x="klasa (class)", y=col, title=col, color='green')
```

##### *widać te relacje, ale chciałbym to mieć na jednym wykresie - i tu muszę też się wspomóc*

```python
df.boxplot(column=['długość kielicha (sepal length)', 
                   'szerokość kielicha (sepal width)', 
                   'długość płatka (petal length)', 
                   'szerokość płatka (petal width)'],
           by='klasa (class)', figsize=(10, 6), color='green')
```

##### *początkowo założyłem, że uda mi się przeprowadzić analizę przy wykorzystaniu jedynie pakietu pandas, ale chcę, żeby te zależności były bardziej widoczne, dlatego dodałem biblioteki `seaborn` i `matplotlib`*

```python
import seaborn as sns
import matplotlib.pyplot as plt

df_melt = df.melt(id_vars='klasa (class)', 
                  var_name='cecha', value_name='wartość')

plt.figure(figsize=(10, 6))
sns.boxplot(data=df_melt, x='cecha', y='wartość', hue='klasa (class)')
plt.xticks(rotation=45)
plt.title("Rozkład cech w zależności od klasy (Iris)")
plt.show()
```

##### *jako uzupełnienie analizy, dodaję korelacje między kolumnami liczbowymi*

```python
korelacje = df.corr(numeric_only=True)
print(korelacje)
```

```python
import matplotlib.pyplot as plt

korelacje = df.corr(numeric_only=True)

plt.figure(figsize=(8, 6))
plt.imshow(korelacje, cmap='coolwarm', interpolation='none')
plt.colorbar()
plt.xticks(range(len(korelacje.columns)), korelacje.columns, rotation=90)
plt.yticks(range(len(korelacje.columns)), korelacje.columns)
plt.title("Macierz korelacji")
plt.tight_layout()
plt.show()
```
